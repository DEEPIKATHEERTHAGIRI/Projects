```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(Study_time,Absences,GPA):
    z = 0.00850395 * Study_time+-0.07000147*Absences+3.23663648*GPA+-5.30639573
    y = sigmoid(z)
    return y
```

```python
Study_time,Absences,GPA=19.833722807854716,7,2.929195591667681
prediction_function(Study_time,Absences,GPA) #1 row=0
```
